```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import pearsonr,linregress
```

```python
mouse_file_path = "Mouse_metadata.csv"
study_file_path = "Study_results.csv"
```

```python
mouse_metadata = pd.read_csv(mouse_file_path)
study_results = pd.read_csv(study_file_path)

study_results.head()
```

```python
#merging both tables (i.e mouse_metdata and study_results) 
combined_df = pd.merge (study_results, mouse_metadata, on = "Mouse ID", how="left")

combined_df
```

```python
#checking the number of mice
num_of_mice = len(combined_df["Mouse ID"].unique())

num_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_rows = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]

duplicate_mice_id = duplicate_rows['Mouse ID'].unique()

duplicate_mice_id 
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combined_df.drop_duplicates('Mouse ID')

clean_df
```

```python
# Checking the number of mice in the clean DataFrame.

num_of_mice_clean_df = len(clean_df)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.


# Calculate statistics
grouped_df_mean = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
grouped_df_median = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
grouped_df_variance = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
grouped_df_std_dev = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
grouped_df_sem =  combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

# Create a DataFrame for the summary statistics
summary_df = pd.DataFrame({
    'Mean Tumor Volume': grouped_df_mean,
    'Median Tumor Volume': grouped_df_median,
    'Tumor Volume Variance': grouped_df_variance,
    'Tumor Volume Std. Dev.': grouped_df_std_dev,
    'Tumor Volume Std. Error': grouped_df_sem 
})

# Display the summary statistics DataFrame
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

summary_stats_df = combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

summary_stats_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint


# Filter the DataFrame
df_filtered = combined_df[combined_df['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Get the last (greatest) timepoint for each mouse
max_timepoints = df_filtered.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_df = pd.merge(max_timepoints, df_filtered, on=['Mouse ID', 'Timepoint'], how='left')

merged_df 
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# loop over treatments
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = merged_df.loc[merged_df['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # add subset to tumor_vol_data
    tumor_vol_data.append(final_tumor_vol)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Create a new figure
plt.figure()

# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol_data, labels=treatments)

# Add title and labels
plt.title('Final Tumor Volume by Treatment')
plt.xlabel('Treatment')
plt.ylabel('Final Tumor Volume (mm3)')

# Show the plot
plt.show()
```

```python
# Filter the DataFrame to include only rows where the 'Drug Regimen' is 'Capomulin'
capomulin_data = combined_df[combined_df['Drug Regimen'] == 'Capomulin']

# Group the data by 'Mouse ID' and calculate the average tumor volume and mouse weight for each mouse
average_data = capomulin_data.groupby('Mouse ID').mean()

# Generate the scatter plot
plt.scatter(average_data['Weight (g)'], average_data['Tumor Volume (mm3)'])

# Add labels and title to the plot
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')

# Display the plot
plt.show()
```

```python
# Calculate the correlation coefficient
correlation = pearsonr(average_data['Weight (g)'], average_data['Tumor Volume (mm3)'])
print(f"The correlation coefficient between mouse weight and average tumor volume is {correlation[0]}")

# Calculate the linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(average_data['Weight (g)'], average_data['Tumor Volume (mm3)'])
print(f"The linear regression model is: y = {slope}x + {intercept}")

# Generate the scatter plot with the linear regression model line
plt.scatter(average_data['Weight (g)'], average_data['Tumor Volume (mm3)'])
plt.plot(average_data['Weight (g)'], slope * average_data['Weight (g)'] + intercept, 'r')

# Add labels and title to the plot
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')

# Display the plot
plt.show()
```
